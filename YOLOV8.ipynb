```python
results=model(img_path)
```

```python
boxes=results[0].boxes
```

```python
boxes
```

```python
img=cv2.imread(img_path)
```

```python
for box in boxes:
  print(box)
  print("*************************************")
```

```python
for box in boxes:
  top_left_x=int(box.xyxy.tolist()[0][0])
  top_left_y=int(box.xyxy.tolist()[0][1])
  bottom_right_x=int(box.xyxy.tolist()[0][2])
  bottom_right_y=int(box.xyxy.tolist()[0][3])

  # cv2.rectangle(img,(top_left_x,top_left_y),(bottom_right_x,bottom_right_y),(50,200,129),2)
  print("top_left_x : "+ str(top_left_x))
  print("top_left_y : "+ str(top_left_y))
  print("bottom_right_x : "+ str(bottom_right_x))
  print("bottom_right_y : "+ str(bottom_right_y))

```

```python
from PIL import Image

# Load the image

image = Image.open(img_path)


# Crop the face region from the image
face_image = image.crop((top_left_x, top_left_y, bottom_right_x, bottom_right_y))


```

```python
import matplotlib.pyplot as plt
%matplotlib inline


# Display the image
plt.imshow(face_image)
plt.axis('off')  # Optional: Turn off axis labels
plt.show()
```
